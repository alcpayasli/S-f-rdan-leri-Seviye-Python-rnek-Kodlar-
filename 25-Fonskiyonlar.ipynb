```python
def carp(*sayılar):
    c = 1
    for i in sayılar:
        c*=i
    print(c)
```

```python
carp(3,4,5)
```

```python
def fonksiyon(**parametreler):
    print(parametreler)
```

```python
fonksiyon(isim = 'Ahmet', soyisim = 'Can', yas = 32)
```

```python
def kayıt_olustur(**kwargs):
    for key, value in kwargs.items():
        print("{:<10}: {}".format(key, value))
```

```python
kayıt_olustur(ad = 'Beren', soyad = 'Can', yas = 12)
```

```python
def bul(*args, **kwargs):
    for kelime in args:
        if kelime in kwargs:
            print("{}..:{}".format(kelime, kwargs[kelime]))
        else:
            print("{} sözlükte yok.".format(kelime))
```

```python
sözlük = {"kitap" : "book", "bilgisayar" : "computer", "programlama": "programming"}
```

```python
bul('kitap', **sözlük)
```

```python
bul('araba', **sözlük)
```

```python
def sayı_uret(bas = 0, bitis = 500, adet = 10):
    sayılar = set()
    import random
    while(len(sayılar)<adet):
        sayılar.add(random.randrange(bas, bitis))
    return sayılar
```

```python
sayı_uret()
```

```python
sayı_uret(bas = 100,adet = 5)
```

```python
for i in range(10):
    print(*sayı_uret())
```

# Global Deyimi

```python
x = 0
def fonk():
    x = 1
    return x
```

```python
print("Fonksiyon dışındaki x değeri:", x)
print("Fonksiyon çağrılınca x değeri:", fonk())
```

```python
x = 10
def fonk():
    x = 20
    print(x)
```

```python
print(x)
fonk()
```

```python
isim = "ali"
def fonk():
    global isim
    isim+=' can'
    print(isim)
```
