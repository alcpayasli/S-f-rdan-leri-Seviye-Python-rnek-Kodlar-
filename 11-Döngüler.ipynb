# while döngüsü

```python
a = 1
while a<10:
    print("{}. döngü".format(a))
    a+=1
```

```python
b = 5
while(b>0):
    print("b'nin değeri: {}".format(b))
    b-=1
```

```python
a = 0
while a<100:
    if a % 5 == 0:
        print(a)
    a+=1
```

```python
a = 0
while True:
    if a % 2 == 0:
        print(a)
    a+=1
    if a>10:
        break # döngüyü keser
```

```python
kelime = 'Merhaba'
sayac = 0
while sayac<len(kelime):
    print(kelime[sayac], end = "-")
    sayac+=1
```

# for döngüsü

```python
harfler = 'abcdef'
for harf in harfler:
    print(harf)
```

```python
sayi = '123456789'
for i in sayi:
    print(i, end = '\t')
```

```python
sayi = '123456789'
for i in sayi:
    if int(i) % 2 == 0:
        print(i, end = '--')
```

```python
yasak_harfler = 'şçöğüİı'
parola = input('Parola belirleyiniz:')
for harf in yasak_harfler:
    if harf in parola:
        print('Parolada {} harfi olamaz'.format(harf))
```

# Döngülerle Hesap Makinesi Örneği

```python
while True:
    print("""

***** UYGULAMAYA HOŞGELDİNİZ *****
İŞLEMLER:
1-TOPLAMA
2-ÇIKARMA
3-ÇARPMA
4-BÖLME
5-ÜS ALMA
0-ÇIKIŞ
""")
    islem = input('İşlemi seçiniz: (Örneğin toplama için 1\'e basınız, çıkmak için 0\'a basınız.)')


    if islem == '0':
        print('Programdan çıkılıyor.')
        break
    else:
        sayi1 = int(input('1.Sayıyı giriniz:'))
        sayi2 = int(input('2.Sayıyı giriniz:'))

        if islem == '1':
            print('Sonuç:', sayi1 + sayi2)
        elif islem == '2':
            print('Sonuç:', sayi1-sayi2)
        elif islem == '3':
            print('Sonuç:', sayi1*sayi2)
        elif islem == '4':
            print('Sonuç:', sayi1/sayi2)
        elif islem == '5':
            print('Sonuç:', sayi1**sayi2)
        else:
            print('Geçersiz işlem girdiniz.')
```

# Döngülerle Parola Örneği

```python
while True:
    parola = input('5 karakterden az olmayan bir parola giriniz:')
    if not parola:
        print('Parola boş bırakılamaz')
    elif len(parola)< 5:
        print('Parola 5 karakterden az olamaz.')
    else:
        print('Parolanız başarı ile oluşturuldu')
        break
```
