# BASİT HESAP MAKİNESİ ÖRNEĞİ

```python
print("""

***** UYGULAMAYA HOŞGELDİNİZ *****
İŞLEMLER:
1-TOPLAMA
2-ÇIKARMA
3-ÇARPMA
4-BÖLME
5-ÜS ALMA
""")
sayi1 = int(input('1.Sayıyı giriniz:'))
sayi2 = int(input('2.Sayıyı giriniz:'))
islem = input('İşlemi seçiniz: (Örneğin toplama için 1\'e basınız)')
if islem == '1':
    print('Sonuç:', sayi1 + sayi2)
elif islem == '2':
    print('Sonuç:', sayi1-sayi2)
elif islem == '3':
    print('Sonuç:', sayi1*sayi2)
elif islem == '4':
    print('Sonuç:', sayi1/sayi2)
elif islem == '5':
    print('Sonuç:', sayi1**sayi2)
else:
    print('Geçersiz işlem girdiniz.')

```
