# % ile biçimlendirme

```python
parola = 'adasdas'
print('Parolanız: %s' % parola)
```

```python
print('%s uygulamalarında %s öncül dildir'%('ML', 'Python'))
```

```python
karakter = 'Python'
for i,j in enumerate(karakter,1):
    print('%s.karakter: %s'% (i,j))
```

```python
for i in range(10):
    print('%s' %i)
```

```python
for i in range(10):
    print('%%%s' %i)
```

```python
print('|%s|'%'Python'.rjust(20))
```

```python
print('|%-20s|' %'Python')
```

```python
print('|%s|'%'Python'.ljust(20))
```

```python
print('Depoda %(miktar)s kilo %(urun)s kaldı'% {'miktar': 20, 'urun': 'elma'})
```

```python
print('Kasada %d kilo elma var.' %32)
```

```python
print('%d' %16.4)
```

```python
print('|%10d|' %65)
```

```python
print('|%-10d|' %65)
```

```python
print('%(sayı)d sayısı bir rakamdır' %{'sayı': 5})
```

```python
print('%05d'%14)
```

```python
print('%.5d'%14)
```

```python
print('%20.5d' %16)
```

```python
print('%010.d' %16)
```

```python
print('%-10.5d|' %16)
```

```python
print('%i sayısının  sekizlik karşılığı: %o' %(10,10))
```

```python
print('%i sayısının  onaltılık karşılığı: %x' %(10,10))
```

```python
print('%i sayısının  onaltılık karşılığı: %X' %(10,10))
```

```python
print('%f' %12.76)
```

```python
print('%.2f' %12.76)
```

```python
print('%.4f' %12.76)
```

```python
print('%c' %'a')
```

```python
print('%c' %59)
```

```python
print('%c' %65)
```

```python
for i in range(128):
    print('%s ==> %c' %(i,i))
```

# format ile biçimlendirme

```python
print('{} kolay bir programlama dilidir.'.format('Python'))
```

```python
print('{1} kolay bir programlama dilidir. Ayrıca {1} esnekliği sayesinde {0} projelerinde kullanılır'.format('Veri Bilimi','Python'))
```

```python
print('{dil} kolay bir programlama dilidir.'.format(dil = 'Python'))
```

```python
print('[{:>15}]'.format('Python'))
```

```python
print('[{:<15}]'.format('Python'))
```

```python
print('[{:^15}]'.format('Python'))
```

```python
for i,j in enumerate(dir('')):
    if i % 3==0:
        print('\n', end = '')
    print('{:<20}'.format(j), end = '')
```

```python
print('{:s}'.format('karakter'))
```

```python
print('{:c}'.format(65))
```

```python
print('{:d}'.format(65))
```

```python
print('{:o}'.format(65))
```

```python
print('{:X}'.format(65))
```

```python
print('{:x}'.format(65))
```

```python
print('{:b}'.format(65))
```

```python
print('{:f}'.format(65))
```

```python
print('{:.2f}'.format(65))
```
