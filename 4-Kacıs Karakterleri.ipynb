# \n (Satır Başı)

```python
print("Birinci satir\nIkinci satir\nUcuncu satir")
```

```python
head = 'Türkiye\'de özgür yazılımın geçmişi'
print(head, "\n", len(head)*'-', sep = '')
```

```python
print('Alışveriş Listesi:', '\n', 18*'-', sep = '')
```
