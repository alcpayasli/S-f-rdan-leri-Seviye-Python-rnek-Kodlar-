```python
sesli_harfler= "aeıioöuü"
sessiz_harfler= "bcçdfgğhjklmnprsştvyz"

kelime = "İstanbul'u dinliyorum gözlerim kapalı."
sesli = ''
sessiz = ''
for i in kelime:
    if i in sesli_harfler and i not in sesli:
        sesli+=i
    elif i in sessiz_harfler and i not in sessiz:
        sessiz+=i
```
