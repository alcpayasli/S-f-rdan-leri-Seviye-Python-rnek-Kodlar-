```python
parola = input('Parola:')
kontrol = True
for i in parola:
    if parola.count(i)>1:
        kontrol = False
if kontrol:
    print('Parola onaylandı')
else:
    print('Aynı karakterden sadece 1 adet kullanabilirsiniz.')
```

```python
kelime = 'Ankara Büyükşehir Belediyesi'
sayac = ''
for i in kelime:
    if not i in sayac:
        sayac+=i
for i in sayac:
    print('{} harfi {} defa geçiyor.'.format(i, kelime.count(i)))
```
