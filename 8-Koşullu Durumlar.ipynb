```python
n = 5
if n>10:
    print("Sayı 10'dan büyüktür.")
```

```python
n = 16
if n>10:
    print("Sayı 10'dan büyüktür.")
if n>15:
    print("Sayı 15'den büyüktür")
if n>20:
    print("Sayı 20'den büyüktür.")
```

```python
sifre = '123456789'
kullanici_girdisi = '1234'
if kullanici_girdisi == sifre:
    print("Parola doğru")
if kullanici_girdisi!=sifre:
    print("Parola hatalı")
```

```python
yas = 20
if yas<18:
    print("Erginlik çağına henüz gelmemişssin")
elif yas>18:
    print("Ergin bir bireysin.")
```

```python
sayi = 40
if sayi<100:
    print("Sayı 100'den küçüktür.")
elif sayi<50:
    print("Sayı 50'den küçüktür.")
elif sayi == 100:
    print("Sayı 100'e eşittir.")
```

# örnek uygulama

```python
kullanici_adi = input('Kullanıcı adını giriniz:')
parola = input('Parolanızı giriniz:')
uzunluk = len(kullanici_adi) + len(parola)
print('Kullanıcı adı ve parolanız toplam {} karakterdir.'.format(uzunluk))
if uzunluk>20:
    print('Kullanıcı adı ve parola 20 karakteri geçmemeli.')
else:
    print('Sisteme hoşgeldiniz.')
```
