```python
for i in range(3):
    parola = input('3 ile 8 karakterden oluşan bir parolanızı giriniz:')
    if not parola:
        print('Parola boş bırakılamaz.')
    elif len(parola) in range(3,9):
        print('Parola oluşturuldu')
        break
    elif i == 2:
        print('3 kez yanlış giriş yapıldı')
    else:
        print("parola 8 karakterden uzun 3 karakterden kısa olmamalı")
```

# pass, break ve continue ifadeleri

```python
while True:
    parola = input('3 ile 8 karakterden oluşan bir parolanızı giriniz:')
    if not parola:
        pass
    elif len(parola) in range(3,9):
        print('Parola oluşturuldu')
        break
    else:
        print("parola 8 karakterden uzun 3 karakterden kısa olmamalı")
```

```python
while 'Python':
    sayi = int(input('Bir sayı giriniz:'))
    if sayi == 0:
        break
    elif sayi<0:
        pass
    else:
        print('Sayi:', sayi)
        break
```

```python
while True:
    sayi = input('en fazla 3 haneli bir sayi giriniz:')
    if sayi == 'iptal':
        break
    elif len(sayi)<=3:
        continue
    print('En fazla 3 haneli sayı girebilirsiniz.')
```

# döngülerde else-break kullanımı

```python
for i in range(5):
    print(i)
else:
    print('Else çalıştı')
```

```python
a = 0
while True:
    a+=1
    print(a)
    if a == 3:
        break
else:
    print('Else çalıştı')
```

# örnek uygulama karakter dizilerini karşılaştırma

```python
ilk_metin= "asdasfddgdhfjfdgdşfkgjdfklgşjdfklgjdfkghdfjghjklsdhajlsdhjkjhkhjjh"
ikinci_metin= "sdfsuıdoryeuıfsjkdfhdjklghjdfklruseldhfjlkdshfljskeeuf"
```

```python
for i in ilk_metin:
    if i not in ikinci_metin:
        print(i)
```

```python
for i in ikinci_metin:
    if i not in ilk_metin:
        print(i)
```

```python
s = ''
for i in ikinci_metin:
    if i not in ilk_metin and i not in s:
        s+=i
print(s)
```

```python
s = ''
print('s\'in ilk id\si:', id(s))
for i in ikinci_metin:
    if i not in ilk_metin and i not in s:
        s+=i
print(s)
print('s\'in son id\si:', id(s))
```
