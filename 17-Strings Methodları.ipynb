# maketrans, translate

```python
kaynak = "şçöğüıŞÇÖĞÜİ"
hedef = "scoguiSCOGUI"
```

```python
tablo = str.maketrans(kaynak,hedef)
```

```python
metin ="Bildiğiniz gibi,internet üzerinde bazen Türkçe karakterleri kullanamıyoruz."
```

```python
metin.translate(tablo)
```

```python
metin ="Bfjflrk öakdhsı yteua idjslyd bdcusldvdj ks?"
q_klavye="qwertyuıopğüasdfghjklşi,zxcvbnmöç."
f_klavye= "fgğıodrnhpqwuieaütkmlyşxjövcçzsb.,"
```

```python
tablo = ' '.maketrans(q_klavye, f_klavye)
metin.translate(tablo)
```

```python
metin = 'Cem Yılmaz'
kaynak = 'CY'
hedef = 'cy'
silinecek = 'eıa'
tablo = str.maketrans(kaynak, hedef, silinecek)
metin.translate(tablo)
```
