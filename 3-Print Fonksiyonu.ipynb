# sep parametresi

```python
print('Python', 'Java', 'C++', sep = '*')
```

```python
print('www', 'google', 'com', sep = '.')
```

```python
print('T', 'C', sep = '.')
```

```python
print('İsim', "Ahmet", sep = ':')
```

```python
print('Python', 'Java', 'C++', sep = None)
```

```python
print("bir", "iki", "üç", "on dört", sep = ' ' + 'mumdur' + ' ')
```

# end parametresi

```python
print('Python')
print('Java')
print('C++')
```

```python
print('Python', end = '-')
print('Java', end = '-')
print('C++')
```

```python
print('Python', end = ' ')
print('Java', end = ' ')
print('C++')
```

```python
print('Birinci satır\nİkinci satır\nÜçüncü satır')
```

```python
print("Birinci satır", "İkinci satır", "Üçüncü satır", sep = '\n')
```

```python
print("bir", "iki", "üç", "on dört", sep = ' ' + 'mumdur' + ' ', end = ' ' + 'mumdur' + " ")
```
