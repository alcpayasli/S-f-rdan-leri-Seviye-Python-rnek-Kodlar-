## Karşılaştırma Operatörleri

```python
parola = 'xyz'
kullanici_parolasi = input('Parolayı giriniz:')
if kullanici_parolasi == parola:
    print('Hoşgeldiniz')
else:
    print('Yanlış parola')
```

```python
sayi = int(input('Bir sayı giriniz:'))
if sayi <=100:
    print('Sayı 100 veya 100\'den küçüktür')
else:
    print('Sayı 100 veya 100\'den büyüktür.')
```

```python
# Not uygulaması
ders_notu = int(input('Ders notunu giriniz:'))
if ders_notu>100 or ders_notu<0:
    print('Böyle bir not olamaz.')
elif ders_notu>=90:
    print('A aldınız')
elif ders_notu>=80:
    print('B aldınız.')
elif ders_notu>=70:
    print('C aldınız.')
else:
    print('Kaldınız.')
```
